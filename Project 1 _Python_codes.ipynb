```python
import pandas as pd
## Load data
air_data = pd.read_csv('AirQuality_Daily_StudentVersion.csv')


## Turn data into a dataframe
df = pd.DataFrame(air_data)

```

## REQUEST 1
### What are the 5 locations in Nebraska with the highest mean and median concentrations of VOC, PM2.5, and PM10?

#### I will:
#### 1. Groupe data by sensor.name
#### 2. Calculate mean and median
#### 3. Sort values
#### 4. Select top 5 locations

```python
voc_summary = air_data.groupby('sensor.name').agg(
    mean_voc=('voc', 'mean'),
    median_voc=('voc', 'median')
).reset_index()

## View summary table
print(voc_summary)
```

```python
## Top 5 locations with highest MEAN VOC
top5_voc_mean = voc_summary.sort_values(by='mean_voc',ascending=False).head(5)

## Top 5 locations with highest MEDIAN VOC
top5_voc_median = voc_summary.sort_values(by='median_voc',ascending=False).head(5)

top5_voc_mean
top5_voc_median
```

```python
pm25_summary = air_data.groupby('sensor.name').agg(
    mean_pm25=('pm2.5_atm', 'mean'),
    median_pm25=('pm2.5_atm', 'median')
).reset_index()

## Sort by mean PM2.5
pm25_top5 = pm25_summary.sort_values(
    by='mean_pm25',
    ascending=False
).head(5)

print("Top 5 Locations – PM2.5 (Mean & Median)")
pm25_top5
```

```python
pm10_summary = air_data.groupby('sensor.name').agg(
    mean_pm10=('pm10.0_atm', 'mean'),
    median_pm10=('pm10.0_atm', 'median')
).reset_index()

## Sort by mean PM10
pm10_top5 = pm10_summary.sort_values(
    by='mean_pm10',
    ascending=False
).head(5)

print("Top 5 Locations – PM10 (Mean & Median)")
pm10_top5
```

## REQUEST 2

### On what days did the maximum values occur and where did these maximums occur?

#### I will find:
#### 1. Maximum VOC
#### 2. Maximum PM2.5
#### 3. Maximum PM10
#### 4. The date and sensor.name where each occurred

```python
## Group by sensor and date, then calculate max VOC for each day/location
voc_daily_max = air_data.groupby(
    ['sensor.name', 'date']
).agg(
    max_voc = ('voc', 'max')
)
## Sort by VOC concentration (highest first)
voc_daily_max_sorted = voc_daily_max.sort_values(
    by = 'max_voc',
    ascending = False
)
## Show the single highest VOC event
voc_max_event = voc_daily_max_sorted.head(1)
print("VOC maximum occurred on:")
voc_max_event
```

```python
## Group by sensor and date, then calculate max PM2.5
pm25_daily_max = air_data.groupby(
    ['sensor.name', 'date']
).agg(
    max_pm25 = ('pm2.5_atm', 'max')
)

## Sort by PM2.5 concentration
pm25_daily_max_sorted = pm25_daily_max.sort_values(
    by = 'max_pm25',
    ascending = False
)
## Show the single highest PM2.5 event
pm25_max_event = pm25_daily_max_sorted.head(1)

print("PM2.5 maximum occurred on:")
pm25_max_event
```

```python
## Group by sensor and date, then calculate max PM10
pm10_daily_max = air_data.groupby(
    ['sensor.name', 'date']
).agg(
    max_pm10 = ('pm10.0_atm', 'max')
)

## Sort by PM10 concentration
pm10_daily_max_sorted = pm10_daily_max.sort_values(
    by = 'max_pm10',
    ascending = False
)

## Show the single highest PM10 event
pm10_max_event = pm10_daily_max_sorted.head(1)

print("PM10 maximum occurred on:")
pm10_max_event
```

## REQUEST 3

### Does humidity and temperature have a noticeable effect on air quality?

#### I will:

#### 1.Create humidity categories

#### 2.Create temperature categories

#### 3.Group data by those categories

#### 4.Compare mean / median VOC, PM2.5, PM10

```python
def humidity_category(hum):
    if hum < 50:
        return "Low"
    elif hum <= 80:
        return "High"
    else:
        return "Very High"

## Apply function to create new column
air_data["humidity_category"] = air_data["humidity"].apply(humidity_category)

air_data.tail()
```

```python
def temp_category(temp):
    if temp < 32:
        return "Below freezing"
    elif 32 < temp <=50:
        return "Cool"
    elif 51 < temp <= 70:
        return "Warm"
    else:
        return "Hot"

## Apply function to create new column
air_data["temp_category"] = air_data["temperature"].apply(temp_category)

air_data.head()
```

## REQUEST 4

### Have there been any AQI health risks (Unhealthy for Sensitive Groups) for PM2.5 and PM10?

#### EPA AQI thresholds used (standard):

#### PM2.5 ≥ 35.5 → Unhealthy for Sensitive Groups

#### PM10 ≥ 155 → Unhealthy for Sensitive Groups

```python
## Filter PM2.5 values that meet AQI "Unhealthy for Sensitive Groups"
unhealthy_pm25 = air_data[air_data['pm2.5_atm'] >= 35.5]

## Select useful columns and sort by date
pm25_risk_events = unhealthy_pm25[
    ['date', 'sensor.name', 'pm2.5_atm']
].sort_values(by='date', ascending=False)

print("PM2.5 AQI Risk Events:")
pm25_risk_events
```

```python
## Filter PM10 values that meet AQI "Unhealthy for Sensitive Groups"
unhealthy_pm10 = air_data[air_data['pm10.0_atm'] >= 155]

## Select useful columns and sort by date
pm10_risk_events = unhealthy_pm10[
    ['date', 'sensor.name', 'pm10.0_atm']
].sort_values(by='date', ascending=False)

print("PM10 AQI Risk Events:")
pm10_risk_events
```

```python
pm10_risk_count = unhealthy_pm10.groupby(
    'sensor.name'
).size().reset_index()

## Rename columns the simple way
pm10_risk_count.columns = ['sensor.name', 'pm10_risk_days']

print("PM10 AQI Risk Days by Location:")
pm10_risk_count
```

```python
pm25_risk_count = unhealthy_pm25.groupby(
    'sensor.name'
).size().reset_index()

## Rename columns the simple way
pm25_risk_count.columns = ['sensor.name', 'pm25_risk_days']

print("PM2.5 AQI Risk Days by Location:")
pm25_risk_count
```

#### These AQI health risk events were likely caused by wildfire smoke, dust from agricultural activities, and local emissions. Calm weather and low wind speeds may have allowed pollutants to build up near the ground.

### BONUS — Does sensor altitude have an impact on air quality values?

```python
altitude_summary = air_data.groupby(
    'sensor.altitude'
).agg(
    mean_voc = ('voc', 'mean'),
    mean_pm25 = ('pm2.5_atm', 'mean'),
    mean_pm10 = ('pm10.0_atm', 'mean')
).reset_index()

## Sort by altitude (highest first)
altitude_summary_sorted = altitude_summary.sort_values(
    by = 'sensor.altitude',
    ascending = False
)

altitude_summary_sorted
```

#### The results suggest that sensor altitude alone does not strongly control air quality. High pollutant concentrations occur at both low and high elevations, indicating that local sources and episodic events likely play a larger role.
